### Análise Final dos resultados ao longo de 2016-2022

```python
import pandas as pd
```

```python
#Importando os dados de cada ano
raw_info_2016 = pd.read_csv('results_df/results_2016.csv', index_col=0)
raw_info_2017 = pd.read_csv('results_df/results_2017.csv', index_col=0)
raw_info_2018 = pd.read_csv('results_df/results_2018.csv', index_col=0)
raw_info_2019 = pd.read_csv('results_df/results_2019.csv', index_col=0)
raw_info_2020 = pd.read_csv('results_df/results_2020.csv', index_col=0)
raw_info_2021 = pd.read_csv('results_df/results_2021.csv', index_col=0)
raw_info_2022 = pd.read_csv('results_df/results_2022.csv', index_col=0)
```

```python
# Copiando cada um dos dataframes para não perder as informações iniciais
info_2016 = raw_info_2016.copy()
info_2017 = raw_info_2017.copy()
info_2018 = raw_info_2018.copy()
info_2019 = raw_info_2019.copy()
info_2020 = raw_info_2020.copy()
info_2021 = raw_info_2021.copy()
info_2022 = raw_info_2022.copy()

# Retorno acumulado de cada coluna para o ano de 2017
for colum in info_2017.columns:
    last_cumulated_2016 = info_2016[colum][-1]
    info_2017[colum] = raw_info_2017[colum]*last_cumulated_2016

# Retorno acumulado de cada coluna para o ano de 2018
for colum in info_2018.columns:
    last_cumulated_2017 = info_2017[colum][-1]
    info_2018[colum] = raw_info_2018[colum]*last_cumulated_2017

# Retorno acumulado de cada coluna para o ano de 2019
for colum in info_2019.columns:
    last_cumulated_2018 = info_2018[colum][-1]
    info_2019[colum] = raw_info_2019[colum]*last_cumulated_2018

# Retorno acumulado de cada coluna para o ano de 2020
for colum in info_2020.columns:
    last_cumulated_2019 = info_2019[colum][-1]
    info_2020[colum] = raw_info_2020[colum]*last_cumulated_2019

# Retorno acumulado de cada coluna para o ano de 2021
for colum in info_2021.columns:
    last_cumulated_2020 = info_2020[colum][-1]
    info_2021[colum] = raw_info_2021[colum]*last_cumulated_2020

# Retorno acumulado de cada coluna para o ano de 2022
for colum in info_2022.columns:
    last_cumulated_2021 = info_2021[colum][-1]
    info_2022[colum] = raw_info_2022[colum]*last_cumulated_2021
```

```python
# Criando um dataframe final
df = pd.concat([info_2016, info_2017, info_2018, 
                    info_2019, info_2020, info_2021,info_2022], axis=0)
df.index = pd.to_datetime(df.index)

display(df)
```

```python
# Plotagem dos resultados para facilitar sua comparação
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6)) 

plt.plot(df.index,df['Cumulative Daily Returns'], label='Estratégia', color ='blue')  
plt.plot(df.index,df['IBOV'], label='IBOV', color ='orange')  
plt.plot(df.index,df['CDI'], label='CDI', color ='green')  

plt.legend()

plt.ylabel('Retorno acumulado')
plt.xlabel('Data')
plt.show()
```
